```python
import pandas as pd
import pyarrow.parquet as pq
import wave
import numpy as np
```

```python
table = pq.read_table('dataset/parpuet_Dataset/train-00000-of-00004.parquet')
df = table.to_pandas()
```

```python
df
```

```python
Mname = []
Mbytes = []
Mcaptions = []
```

```python
for iteam in range(len(df)):
    data = df.iloc[iteam,:]
    Mname.append(data[0]['path'])
    Mbytes.append(data[0]['bytes'])
    Mcaptions.append(data[1])
    
```

```python
Mname
```

```python
Mbytes
```

```python
Mcaptions
```

```python
def bytes_to_wav(bytes_data, output_file, sample_rate=44100):
    audio_array = np.frombuffer(bytes_data, dtype=np.int16)

    with wave.open(f'dataset/music/{output_file}', 'wb') as wav_file:
        wav_file.setnchannels(1)  # Assuming mono audio
        wav_file.setsampwidth(2)   # 2 bytes per sample (16-bit)
        wav_file.setframerate(sample_rate)
        wav_file.writeframes(audio_array)
```

```python
for i in range(len(df)):
    bytes_to_wav(Mbytes[i],Mname[i])
```

```python
my_dict = {
    'audio' : Mname,
    'caption' : Mcaptions
}

df_MNC = pd.DataFrame(my_dict)
df_MNC.head()
```

```python
df_MNC.to_csv('Mname&Captions.csv', index=False)
```
